```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Academic.csv')
df
```

```python
df.isnull()
```

```python
series=pd.isnull(df['Che_marks'])
df[series]
```

```python
df.notnull()
```

```python
series1=pd.notnull(df['Che_marks'])
df[series1]
```

```python
df['Che_marks'] = df['Che_marks'].fillna(df['Che_marks'].mean())
df['Phy_marks'] = df['Phy_marks'].fillna(df['Phy_marks'].median())
df['EM1_marks'] = df['EM1_marks'].fillna(df['EM1_marks'].min())
df['PPS_marks'] = df['PPS_marks'].fillna(df['PPS_marks'].max())
df['SME_marks'] = df['SME_marks'].fillna(df['SME_marks'].std())
df
```
